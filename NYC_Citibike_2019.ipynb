```python
nycbike_2019dec = pd.read_csv(filepath_dec19)
nycbike_2019nov = pd.read_csv(filepath_nov19)
nycbike_2019oct = pd.read_csv(filepath_oct19)
nycbike_2019sep = pd.read_csv(filepath_sep19)
nycbike_2019aug = pd.read_csv(filepath_aug19)
nycbike_2019jul = pd.read_csv(filepath_jul19)
nycbike_2019jun = pd.read_csv(filepath_jun19)
nycbike_2019may = pd.read_csv(filepath_may19)
nycbike_2019apr = pd.read_csv(filepath_apr19)
nycbike_2019mar = pd.read_csv(filepath_mar19)
nycbike_2019feb = pd.read_csv(filepath_feb19)
nycbike_2019jan = pd.read_csv(filepath_jan19)
```

```python
nycbike_2019dec.head(10)
```

```python
nycbike_2019 = pd.concat([nycbike_2019dec,nycbike_2019nov,nycbike_2019oct,nycbike_2019sep,nycbike_2019aug,nycbike_2019jul,
                          nycbike_2019jun,nycbike_2019may,nycbike_2019apr,nycbike_2019mar,nycbike_2019feb,nycbike_2019jan])
```

```python
nycbike_2019.rename(columns={"tripduration": "Trip Duration",
                                            "starttime": "Start Time",
                                            "stoptime": "Stop Time",
                                            "start station id": "Start Station ID",
                                            "start station name": "Start Station Name",
                                            "start station latitude": "Start Station Latitude",
                                            "start station longitude": "Start Station Longitude",
                                            "end station id": "End Station ID",
                                            "end station name": "End Station Name",
                                            "end station latitude": "End Station Latitude",
                                            "end station longitude": "End Station Longitude",
                                            "bikeid": "Bike ID",
                                            "usertype": "User Type",
                                            "birth year": "Birth Year",
                                            "gender": "Gender"
                                           }, inplace = True)
nycbike_2019.head()
```
