```python
nycbike_2020dec = pd.read_csv(filepath_dec20)
nycbike_2020nov = pd.read_csv(filepath_nov20)
nycbike_2020oct = pd.read_csv(filepath_oct20)
nycbike_2020sep = pd.read_csv(filepath_sep20)
nycbike_2020aug = pd.read_csv(filepath_aug20)
nycbike_2020jul = pd.read_csv(filepath_jul20)
nycbike_2020jun = pd.read_csv(filepath_jun20)
nycbike_2020may = pd.read_csv(filepath_may20)
nycbike_2020apr = pd.read_csv(filepath_apr20)
nycbike_2020mar = pd.read_csv(filepath_mar20)
nycbike_2020feb = pd.read_csv(filepath_feb20)
nycbike_2020jan = pd.read_csv(filepath_jan20)
```

```python
nycbike_2020dec.head(10)
```

```python
nycbike_2020 = pd.concat([nycbike_2020dec,nycbike_2020nov,nycbike_2020oct,nycbike_2020sep,nycbike_2020aug,nycbike_2020jul,
                          nycbike_2020jun,nycbike_2020may,nycbike_2020apr,nycbike_2020mar,nycbike_2020feb,nycbike_2020jan])
```

```python
nycbike_2020.rename(columns={"tripduration": "Trip Duration",
                                            "starttime": "Start Time",
                                            "stoptime": "Stop Time",
                                            "start station id": "Start Station ID",
                                            "start station name": "Start Station Name",
                                            "start station latitude": "Start Station Latitude",
                                            "start station longitude": "Start Station Longitude",
                                            "end station id": "End Station ID",
                                            "end station name": "End Station Name",
                                            "end station latitude": "End Station Latitude",
                                            "end station longitude": "End Station Longitude",
                                            "bikeid": "Bike ID",
                                            "usertype": "User Type",
                                            "birth year": "Birth Year",
                                            "gender": "Gender"
                                           }, inplace = True)
nycbike_2020.head()
```

```python
nycbike_2020["Start Time"] = pd.to_datetime(nycbike_2020["Start Time"])
nycbike_2020["Stop Time"] = pd.to_datetime(nycbike_2020["Stop Time"])
nycbike_2020.head()
```
